# Mem-wise

```python
import os
import numpy as np
import matplotlib.pyplot as plt

def read_npy_files(directory):
    data = {}
    for filename in os.listdir(directory):
        if filename.endswith(".npy"):
            filepath = os.path.join(directory, filename)
            data[filename] = np.load(filepath)
    return data

def plot_accuracies(data):
    plt.figure(figsize=(10, 6))
    for filename, accuracies in data.items():
        plt.plot(accuracies, label=filename)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Model Accuracies over Epochs')
    plt.legend()
    plt.grid(True)
    plt.show()

# 目录路径
directory = 'acc/cifar100/'

# 读取 npy 文件
data = read_npy_files(directory)

# 绘制图表
plot_accuracies(data)
```
